## Project Goal

The primary goal of this project is to empower marketing teams to quickly assess campaign effectiveness, identify top-performing segments, and make data-driven decisions to optimize future spending and maximize Return on Investment (ROI). ## Features

* **Data Cleaning & Preparation:** Handles missing values and standardizes column names for consistent analysis.

* **Segmentation Analysis:** Provides in-depth analysis of ROI segmented by various dimensions like campaign, age group, gender, and advertising platform.
* **Visualization:** Generates insightful charts and plots to visually represent key performance indicators and trends.
* **Correlation Analysis:** Identifies which underlying metrics most influence overall campaign ROI, helping to pinpoint performance drivers.
* **Export Capabilities:** Outputs summary tables and charts in common formats (CSV, PNG) for easy sharing with stakeholders.
